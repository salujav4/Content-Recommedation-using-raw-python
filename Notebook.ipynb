```python
content['episodes'] = content['episode_count']*content['season_count']
#Merging the season and episode column to determine the total length of the content
```

```python
content = content.drop(['episode_count','season_count'], axis=1)
#Dropping the extra columns
```

```python
for i in range(len(content['episodes'])):
    if content['episodes'][i]<101:
        content['episodes'][i] = 1
    elif content['episodes'][i]>100 and content['episodes'][i]<301:
        content['episodes'][i] = 2
    elif content['episodes'][i]>300 and content['episodes'][i]<501:
        content['episodes'][i] = 3
    else:
        content['episodes'][i] = 4
        
#Converting the new episodes column into categorical values to classify the content
```

```python
test = pd.read_csv('test.csv')
#Loading the test as these will main users that we will be working on
```

```python
index = []
for i in user['user_id']:
    if i not in list(test['user_id']):
        index.append(user[user['user_id'] == i].index[0])
#Copying the indexes of users from the user dataframe which are in the test dataframe
```

```python
final = user.drop(index).reset_index(drop=True)
```

```python
final
```

```python
final_content = []
for i in range(len(final)):
    total_content = []
    for j in relation[relation['user_id']==final['user_id'][i]]['content_id']:
        a = content[content['content_id'] == j]['content_type']
        total_content.append(a)
        sum = 0
    for k in pd.DataFrame(np.array(total_content)).value_counts():
        sum = sum + k
    l = pd.DataFrame(np.array(total_content)).value_counts().to_dict()
    for m in range(len(l)):
        if len(l)>=0:
            l[list(l.keys())[m]] = (l[list(l.keys())[m]]/sum)*10
        elif len(l)==0:
            l[list(l.keys())[m]] = 0.33*10
    final_content.append(l)
    
#Getting the content that a user watches and keeping it on a ratio with total content watched and 
#multiplied by 10 to get the content proportionate 10
```

```python
final = final.join(pd.DataFrame(final_content))
```

```python
final = final.fillna(0)
```

```python
final
```

```python
final_lang = []
for i in range(len(final)):
    total_content = []
    for j in relation[relation['user_id']==final['user_id'][i]]['content_id']:
        a = content[content['content_id'] == j]['language']
        total_content.append(a)
        sum = 0
    for k in pd.DataFrame(np.array(total_content)).value_counts():
        sum = sum + k
    l = pd.DataFrame(np.array(total_content)).value_counts().to_dict()
    for m in range(len(l)):
        if len(l)>=0:
            l[list(l.keys())[m]] = (l[list(l.keys())[m]]/sum)*10
        elif len(l)==0:
            l[list(l.keys())[m]] = 0.33*10
    final_lang.append(l)
#Getting the language which a user watches and keeping it on a ratio with total languages watched
#and multiplied by 10 to get the language proportionate to 10
```

```python
final = final.join(pd.DataFrame(final_lang))
```

```python
final = final.fillna(0)
```

```python
final_genre = []
for i in range(len(final)):
    total_content = []
    for j in relation[relation['user_id']==final['user_id'][i]]['content_id']:
        a = content[content['content_id'] == j]['genre']
        total_content.append(a)
        sum = 0
    for k in pd.DataFrame(np.array(total_content)).value_counts():
        sum = sum + k
    l = pd.DataFrame(np.array(total_content)).value_counts().to_dict()
    for m in range(len(l)):
        if len(l)>=0:
            l[list(l.keys())[m]] = (l[list(l.keys())[m]]/sum)*10
        elif len(l)==0:
            l[list(l.keys())[m]] = 0.33*10
    final_genre.append(l)
    
#Getting the genre which a user watches and keeping it on a ratio with total genres watched
#and multiplied by 10 to get the genres proportionate to 10
```

```python
final = final.join(pd.DataFrame(final_genre).fillna(0))
```

```python
short_duration = []
long_duration = []
for i in range(len(final)):
    user_duration_short = []
    user_duration_long = []
    for j in relation[relation['user_id'] == final['user_id'][i]]['content_id']:
        rating = content[content['content_id'] == j]['duration']
        if int(rating)<int(content['duration'].mode()):
            user_duration_short.append(rating)
        else:
            user_duration_long.append(rating)
    try:
        short_duration.append(len(user_duration_short)/(len(user_duration_short)+len(user_duration_long)))
        long_duration.append(len(user_duration_long)/(len(user_duration_short)+len(user_duration_long)))
    except ZeroDivisionError:
        short_duration.append(0)
        long_duration.append(0)
    
#Dividing the duration of content to short and long and then counting the number of content watched
#by user in proportionate to 10
```

```python
final['Short_duration'] = short_duration
final['Long_duration'] = long_duration
```

```python
final[['Short_duration','Long_duration']] = final[['Short_duration','Long_duration']]*10
```

```python
final.iloc[:,5:] = np.round(final.iloc[:,5:])
```

```python
total_length = []
for i in range(len(final)):
    user_length = []
    for j in relation[relation['user_id'] == final['user_id'][i]]['content_id']:
        user_length.append(content[content['content_id'] == j]['episodes'])
    total_length.append(pd.DataFrame(np.array(user_length)).value_counts().to_dict())
    
#Converting the length into groups of 200 episodes and creating 4 categories and counting the
#number of content in each category.
```

```python
final = final.join(pd.DataFrame(total_length).fillna(0))
```

```python
content_type = list(final.iloc[:,5:9].columns)
language = list(final.iloc[:,9:20].columns)
genre = list(final.iloc[:,20:42].columns)
duration = list(final.iloc[:,42:44].columns)
length = list(final.iloc[:,44:])

#Dividing all the identifiers created above and converting them into lists
```

```python
for i in range(len(length)):
    length[i] = int(length[i])
```

```python
highest_rating = content[content['rating']==10]
final_list = []
for user in final['user_id']:
    counter = []
    final_user_list = []
    usersss = []
    content_for_user = []
    watched = list(relation[relation['user_id'] == 'user_12318@domain.com']['content_id'])
    if len(watched) > 0:
        for i in content_type:
            if final[i][0]>0:
                sorted_content = highest_rating[highest_rating['content_type']==i]
                for j in language:
                    if final[j][0]>0:
                        sorted_language = sorted_content[sorted_content['language']==j]
                        for k in genre:
                            if final[k][0]>0:
                                sorted_genre = sorted_language[sorted_language['genre']==k]
                                for l in duration:
                                    if final[l][0]>0:
                                        sorted_duration = sorted_genre[sorted_genre['duration']>3240000]
                                        for m in length:
                                            if final[m][0]>0:
                                                sorted_length = sorted_duration[sorted_duration['episodes']==m]
                                                for a in range(int(np.round((final[i][0]+final[j][0]+final[k][0]+final[l][0]+final[m][0])/5))):
                                                    if set([i,j,k,l,m]) not in counter:
                                                        try:
                                                            usersss.append(random.choice(list(sorted_length['content_id'])))
                                                            counter.append(set([i,j,k,l,m]))
                                                        except:
                                                            pass
                                                    
    unique=[]
    for x in usersss:
        if x not in unique:
            unique.append(x)                                                        

    final_for_user = []
    for k in unique:
        if k not in watched:
            final_for_user.append(k)
        
    for w in range(10):
        final_user_list.append(random.choice(final_for_user))
    final_list.append(final_user_list)
    
#This syntax is created to convert a list of 10 recommended content for each user.
#First all the permutations are created using loops for all identifiers with the concept of set is
#used to keep it on-repetitive. There area ifs, the first is for old users who are
#recommended highest rated content of their type with the help of identifiers.
#The second is for the new users who have not watched any content on the platform who are recommended
#highest rated content of all types. So that they can create their preferences.
#Then for the old users, the recommended content who the user has already watched are eliminated
#At the end the content is recommended using random.choice from the remaining recommended content.
```
